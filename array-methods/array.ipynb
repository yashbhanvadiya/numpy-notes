```python
import numpy as np
import time
```

Create array from list

```python
arr_1d = np.array([1, 2, 3, 4, 5])
print("arr_1d:", arr_1d)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("arr_2d:", arr_2d)
```

List vs Numpy array

```python
py_list = [1, 2, 3]
print("Python list multiplication:", py_list * 2)

np_array = np.array([1, 2, 3])
print("Numpy list multiplication:", np_array * 2)

start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List operation time:", time.time() - start)

start = time.time()
np_array = np.arange(1000000) * 2
print("\n Numpy operation time:", time.time() - start)
```

Create array from scratch

```python
zeros = np.zeros((3, 4))
print("zeros array \n", zeros)

ones = np.ones((3, 4))
print("ones array \n", ones)

full = np.full((2, 5), 7)
print("full array \n", full)

random = np.random.random([2, 4])
print("random array \n", random)

sequence = np.arange(0, 10, 2)
print("sequence array \n", sequence)
```

Vector, Matrix and Tensor

```python
vector = np.array([1, 2, 3])
print("vector", vector)

matrix = np.array([
                    [1, 2, 3], 
                    [4, 5, 6]
                ])
print("matrix", matrix)

tensor = np.array([[[1, 2], [3, 4]], 
                    [[5, 6], [7, 8]]])
print("tensor", tensor)
```

Array Properties

```python
arr = np.array([
                    [1, 2, 3], 
                    [4, 5, 6]
                ])
print("Shape", arr.shape)
print("Dimension", arr.ndim)
print("Size", arr.size)
print("DType", arr.dtype)
```

Array Reshaping

```python
arr = np.arange(12)
print("original array", arr)

reshaped = arr.reshape((3, 4))
print("\n reshaped array", reshaped)

flattend = reshaped.flatten()
print("\n flattend array", flattend)

## Reveled - return view instead of copy
reveled = reshaped.flatten()
print("\n reveled array", reveled)

## Transpose
transpose = reshaped.T
print("\n transpose array", transpose)
```
