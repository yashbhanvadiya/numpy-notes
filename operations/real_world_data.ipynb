Advance Operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Restaurant: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 15000, 18000, 22000, 25000], # Tulsi Restaurant
    [2, 12000, 14000, 16000, 19000], # McDonald's
    [3, 20000, 22000, 26000, 30000], # Burger King
    [4, 18000, 21000, 28000, 27000], # Domino's
    [5, 35000, 18500, 20500, 23000]  # Dolly Chai wallah
])
```

```python
print("Sales data shape", sales_data.shape)
print("Sample data for first 3 restaurant", sales_data[:, 1:])
```

```python
## Total sales per year
print(np.sum(sales_data, axis=0))

year_total = np.sum(sales_data[:, 1:], axis=0)
print("year_total", year_total)

## Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print("min_sales", min_sales)

## Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print("max_sales", max_sales)

## Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("avg_sales", avg_sales)

## Cumulative sales per restaurant
cumsum_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("cumsum_sales", cumsum_sales)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(np.mean(cumsum_sales, axis=0))
plt.title("Avrage cumulative sales accross all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vectors addition", vector1 + vector2)

print("\n Multiplication vectors", vector1 * vector2)

print("\n Dot product", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\n Angle", angle)
```
